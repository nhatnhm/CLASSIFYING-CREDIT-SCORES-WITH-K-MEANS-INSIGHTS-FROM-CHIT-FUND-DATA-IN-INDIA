**Table cf1_delhi_collateral**

```python
# --- Loading the dataset ---

# Define the file path for the dataset
dataset_path = r"Data\cf1_delhi_collateral.csv"

# Load the dataset into a pandas DataFrame
cf1_delhi_collateral = pd.read_csv(dataset_path)

# Display the DataFrame to verify successful loading
cf1_delhi_collateral
```

```python
# --- Selecting necessary columns from the dataset ---

# Create a new DataFrame with only the required columns
df_cf1_delhi_collateral = cf1_delhi_collateral[
    [
        # 'Unnamed: 0',  # This column is commented out as it is not required
        'chit_id',        # Unique ID for each chit
        'p_recno',        # Record number of participants
        'collateral',     # Collateral value or information
        # 'chit_value',   # Additional columns are commented out (not required here)
        # 'duration', 
        # 'monthly_contribution', 
        # 'year'
    ]
].copy()  # Use `.copy()` to avoid modifying the original DataFrame
```

```python
# --- Adding a new column: n_collateral ---

# Calculate the number of collaterals for each chit_id
df_cf1_delhi_collateral['n_collateral'] = (
    df_cf1_delhi_collateral
    .groupby('chit_id')['collateral']
    .transform('count')  # Count the number of entries per chit_id
)

# The new column 'n_collateral' indicates the total collateral count for each chit_id
```

```python
# --- Cleaning the DataFrame ---

# Remove the 'collateral' column and keep only the necessary columns
df_cf1_delhi_collateral = (
    df_cf1_delhi_collateral[
        [
            'chit_id',       # Unique ID for each chit
            'p_recno',       # Record number of participants
            'n_collateral',  # Number of collaterals for each chit
            # Additional columns are commented out (not used in this step):
            # 'chit_value', 'duration', 'monthly_contribution', 'year'
        ]
    ]
    .drop_duplicates()  # Remove duplicate rows
    .copy()             # Create a copy to avoid altering the original data
)

# Now, df_cf1_delhi_collateral contains only unique rows with the required columns.
```

```python
# --- Creating a final copy of the cleaned DataFrame ---

# Create a final copy of the cleaned DataFrame for further analysis or processing
df_cf1_delhi_collateral_final = df_cf1_delhi_collateral.copy()

# Display the final DataFrame
df_cf1_delhi_collateral_final
```

**Table cf1_delhi_surety**

```python
# --- Loading the second dataset ---

# Define the file path for the second dataset
dataset_path2 = r"Data\cf1_delhi_surety.csv"

# Load the dataset into a pandas DataFrame
cf1_delhi_surety = pd.read_csv(dataset_path2)

# Display the final DataFrame to verify its contents
cf1_delhi_surety
```

```python
# --- Cleaning and preparing the cf1_delhi_surety dataset ---

# Step 1: Keep only the necessary columns
cf1_delhi_surety = cf1_delhi_surety[
    [
        # 'Unnamed: 0',  # Commented out as it is not required
        'chit',           # Name or ID of the chit
        'chit_id',        # Unique chit identifier
        'p_recno',        # Record number of participants
        # 'winning_aucn',  # Commented columns not needed here
        # 'duration',
        # 'fman',
        # 'surety_p_recno',
        'n_surety',       # Number of sureties associated with each chit
        # 'age', 'salary', 'sex', 'occupation', 
        # 'other_chits', 'surety_others', 'years_of_service',
        # 'house_owner', 'land_owner', 'income_tax', 'insurance_policy',
        # 'insurance_amount', 'res_pin', 'off_pin', 'chit_value',
        # 'monthly_contribution', 'year'
    ]
].copy()  # Create a new DataFrame copy to avoid modifying the original

# Step 2: Replace missing values in the 'n_surety' column with 0
cf1_delhi_surety['n_surety'] = cf1_delhi_surety['n_surety'].fillna(0)

# Step 3: Remove duplicate rows
cf1_delhi_surety = cf1_delhi_surety.drop_duplicates()
```

```python
# --- Creating a final copy of the cleaned cf1_delhi_surety dataset ---

# Create a final copy of the cleaned DataFrame for further analysis or processing
cf1_delhi_surety_final = cf1_delhi_surety.copy()

# Display the final DataFrame to verify its contents
cf1_delhi_surety_final
```

**Table cf1_delhi_transaction_data**

```python
# --- Loading the third dataset ---

# Define the file path for the transaction data
dataset_path3 = r"Data\cf1_delhi_transaction_data.csv"

# Load the dataset into a pandas DataFrame
cf1_delhi_transaction_data = pd.read_csv(dataset_path3)

# Display the dataset to verify successful loading
cf1_delhi_transaction_data
```

```python
# --- Cleaning and preparing the transaction dataset ---

# Select only the necessary columns from the transaction dataset
cf1_delhi_transaction_data = cf1_delhi_transaction_data[
    [
        # 'Unnamed: 0',  # Commented out columns not needed in this step
        'chit',                   # Name or ID of the chit
        'chit_id',                # Unique chit identifier
        'p_recno',                # Participant record number
        # 'aucn_no',
        'aucn_date',              # Auction date
        'inst_due',               # Installment due
        'inst_paid',              # Installment paid
        'inst_spread',            # Spread of installments
        'total_inst_due',         # Total installment due
        'total_inst_paid',        # Total installment paid
        'div_due',                # Dividend due
        'div_paid',               # Dividend paid
        'total_div_due',          # Total dividend due
        'total_div_paid',         # Total dividend paid
        'participation',          # Participation indicator
        'all_bids',               # Total bids
        'win_loss',               # Win/loss status
        'win_bid_amt',            # Winning bid amount
        'prized_amt',             # Prized amount
        'chit_value',             # Value of the chit
        'start_date',             # Start date of the chit
        'monthly_contribution',   # Monthly contribution amount
        'duration',               # Duration of the chit
        'month',                  # Month indicator
        # 'tot_memb', 'fman_tkt',
        # 'bylaw_no',
        'penalty',                # Penalty amount
        # 'postage_cost', 'nj_stamp_cost', 'other_cost',
        'by_chq',                 # Payments made by cheque
        'by_cash',                # Payments made in cash
        'by_other',               # Payments made by other methods
        'bounced_chq',            # Bounced cheque count
        # 'last_trans_date',
        'last_payment_date',      # Date of the last payment
        'missed_inst',            # Number of missed installments
        'missed_div',             # Number of missed dividends
        'diff_inst',              # Difference in installments
        'no_trans',               # Number of transactions
        'total_trans',            # Total number of transactions
        'multi_payment',          # Indicator for multiple payments
        'early_payment',          # Indicator for early payments
        'part_payment',           # Indicator for partial payments
        'irr_payment',            # Indicator for irregular payments
        'late_payment',           # Indicator for late payments
        'default',                # Default indicator
        'monthly_income',         # Monthly income of the participant
        'sex',                    # Gender of the participant
        'age',                    # Age of the participant
        'occupation',             # Occupation of the participant
        'lottery',                # Lottery participation indicator
        # 'bid_type',
        'before_after',           # Indicator for before/after status
        # 'all_trans',
        'default_90'              # Default within 90 days
    ]
].copy()  # Create a copy of the cleaned DataFrame
```

```python
# --- Regenerating data for the "bid_type" column ---

# Step 1: Generate "bid_type" based on conditions
# - Count the number of participants per "chit" and "month" where:
#   - "participation" equals 1 (active participation)
#   - "all_bids" is not equal to 0 (at least one bid is present)
cf1_delhi_transaction_data['bid_type'] = cf1_delhi_transaction_data[
    (cf1_delhi_transaction_data['participation'] == 1) & 
    (cf1_delhi_transaction_data['all_bids'] != 0)
].groupby(['chit', 'month'])['participation'].transform('count')

# Step 2: Replace NaN values in the "bid_type" column with 0
cf1_delhi_transaction_data['bid_type'] = cf1_delhi_transaction_data['bid_type'].fillna(0)

# Step 3: Apply a transformation to "bid_type":
# - If the value is greater than 2, assign it a value of 2
cf1_delhi_transaction_data['bid_type'] = cf1_delhi_transaction_data['bid_type'].apply(lambda x: 2 if x > 2 else x)

# Step 4: Count the unique values in the "bid_type" column
unique = cf1_delhi_transaction_data['bid_type'].value_counts()

# Display the unique value counts for verification
unique
```

```python
# --- Finalizing the transaction dataset ---

# Step 1: Create a final copy of the cleaned and processed transaction dataset
cf1_delhi_transaction_data_final = cf1_delhi_transaction_data.copy()

# Step 2: Display the final dataset to confirm its structure and content
cf1_delhi_transaction_data_final
```

**Merge the three tables into the final dataframe**

```python
# --- Merging Collateral and Surety Tables ---

# Step 1: Merge `cf1_delhi_surety_final` with `df_cf1_delhi_collateral_final`
# - Join on the columns `chit_id` and `p_recno` using an outer join
merge_collateral_surety = pd.merge(
    cf1_delhi_surety_final, 
    df_cf1_delhi_collateral_final, 
    on=['chit_id', 'p_recno'], 
    how='outer'
)

# Step 2: Drop the `p_recno` column as it is no longer needed
merge_collateral_surety = merge_collateral_surety.drop(columns=['p_recno'])

# Step 3: Create a final copy of the merged dataset
merge_collateral_surety_final = merge_collateral_surety.copy()

# Step 4: Display the final merged dataset for verification
merge_collateral_surety_final
```

**Final dataframe**

```python
# --- Merging Transaction Data with Collateral & Surety Data ---

# Step 1: Merge the transaction data (`cf1_delhi_transaction_data_final`) 
#         with the merged collateral and surety data (`merge_collateral_surety_final`).
# - Join on the columns `chit` and `chit_id` using a left join.
df = pd.merge(
    cf1_delhi_transaction_data_final, 
    merge_collateral_surety_final, 
    on=['chit', 'chit_id'], 
    how='left'
)

# Step 2: Display the final DataFrame for verification
df
```

# Data Modeling

## Finding The Optimal Number Of Clusters

###  Elbow method

```python
# --- Using Elbow Method to Determine Optimal K for KMeans Clustering ---

# Ignore FutureWarnings
warnings.filterwarnings("ignore", category=FutureWarning)

# Initialize an empty list to store the loss values for each K
losses = []
K = 10  # Max number of clusters to test

# Loop over the range of possible K values
for i in range(1, K):
    # 1. Train the KMeans model with number of clusters = i
    kmeans_i = KMeans(n_clusters=i, init='k-means++', random_state=10)
    kmeans_i.fit(df_scale)
    
    # 2. Calculate the distortion (loss)
    # 2.1. Calculate the distance to all centroids
    d2centroids = cdist(df_scale, kmeans_i.cluster_centers_, 'euclidean')  # shape (n, k)
    
    # 2.2. Get the minimum distance to the nearest centroid
    min_distance = np.min(d2centroids, axis=1)  # shape (n)
    
    # Sum the minimum distances to calculate the loss
    loss = np.sum(min_distance)
    
    # Append the loss to the list
    losses.append(loss)

# Plot the Elbow Method to visualize the distortion values
plt.figure(figsize=(12, 8))
plt.plot(np.arange(1, K), losses, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion', fontsize=16 , loc = 'left', fontweight='bold')

# Save Image
plt.savefig("./Image/elbow_method_distortion.png", 
            dpi = 300,
            pad_inches = 1)

plt.show()
```

###  Silhouette  method

```python
# --- Silhouette Analysis for KMeans Clustering (2-7 Clusters) using Yellowbrick ---

# Create a subplot with 3 rows and 2 columns
fig, ax = plt.subplots(3, 2, figsize=(13,8))
fig.suptitle('Silhouette Analysis for 2-7 Clusters', 
             size=16, 
             ha = 'left', # Adjusts the horizontal position of the title
             fontweight='bold', 
             x = 0.01 # Aligns the title to the left
            ) 
plt.tight_layout()

# Loop through the number of clusters from 2 to 7
for i in [2, 3, 4, 5, 6, 7]:
    '''
    Create KMeans instance with the specified number of clusters
    '''
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=42)
    
    # Determine the position of subplots
    q, mod = divmod(i, 2)
    
    '''
    Create SilhouetteVisualizer instance with the KMeans instance
    Fit the visualizer to the data
    '''
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(df_scale)

# Save Image
# plt.savefig("./Image/silhouette_analysis.png", 
#             dpi = 300,
#             pad_inches = 1)

# Display the silhouette plot
plt.show()
```
